```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

# Bubble Sort

```python
bubble = pd.read_csv("bubble.csv")
bubble
```

```python
plt.figure(figsize=(8, 6))
plt.plot(
    bubble["Input(n)"],
    bubble[" count1"],
    label="slope",
    color="blue",
    linewidth=1,
    marker="o",
    markersize=3,
)
plt.title("GRAPH", fontdict={"fontname": "courier new", "fontsize": 20})
plt.xlabel("N", fontdict={"fontname": "courier new", "fontsize": 10})
plt.ylabel("COUNT", fontdict={"fontname": "courier new", "fontsize": 10})
plt.legend()
plt.savefig("BubbleSortGraph.png")
plt.show()
```
